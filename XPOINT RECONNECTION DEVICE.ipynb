```python
# Figure 1: Experimental Apparatus
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Neutrino Detector
detector_path = '../01_Experimental_Data/field_measurements/Neutrino Detector.png'
if os.path.exists(detector_path):
    img_detector = mpimg.imread(detector_path)
    axes[0].imshow(img_detector)
    axes[0].set_title('Neutrino Detector - Field Measurement Apparatus', fontsize=12, fontweight='bold')
    axes[0].axis('off')
else:
    axes[0].text(0.5, 0.5, 'Image not found', ha='center')

# Neutrino Trap
trap_path = '../01_Experimental_Data/field_measurements/Neutrino Trap.png'
if os.path.exists(trap_path):
    img_trap = mpimg.imread(trap_path)
    axes[1].imshow(img_trap)
    axes[1].set_title('Neutrino Trap - Detection Configuration', fontsize=12, fontweight='bold')
    axes[1].axis('off')
else:
    axes[1].text(0.5, 0.5, 'Image not found', ha='center')

plt.tight_layout()
plt.suptitle('Figure 1: X-Point Detection Apparatus', fontsize=14, fontweight='bold', y=1.02)
plt.show()
```

```python
# Figure 2: Observed X-Point Events
fig, axes = plt.subplots(1, 2, figsize=(20, 8))

# First Contact
first_path = '../01_Experimental_Data/xpoint_simulation_runs/a FirstContact.jpg'
if os.path.exists(first_path):
    img_first = mpimg.imread(first_path)
    axes[0].imshow(img_first)
    axes[0].set_title('Event 1: "First Contact" - Initial X-Point Formation', fontsize=14, fontweight='bold')
    axes[0].axis('off')
    # Add annotation
    axes[0].text(0.02, 0.98, 'X-Point Region', transform=axes[0].transAxes, 
                fontsize=12, color='yellow', backgroundcolor='black',
                verticalalignment='top', bbox=dict(boxstyle='round', facecolor='black', alpha=0.7))

# Transporter
trans_path = '../01_Experimental_Data/xpoint_simulation_runs/a Transporter.jpg'
if os.path.exists(trans_path):
    img_trans = mpimg.imread(trans_path)
    axes[1].imshow(img_trans)
    axes[1].set_title('Event 2: "Transporter" - Sustained X-Point Configuration', fontsize=14, fontweight='bold')
    axes[1].axis('off')
    # Add annotation
    axes[1].text(0.02, 0.98, 'Energy Transfer Region', transform=axes[1].transAxes, 
                fontsize=12, color='yellow', backgroundcolor='black',
                verticalalignment='top', bbox=dict(boxstyle='round', facecolor='black', alpha=0.7))

plt.tight_layout()
plt.suptitle('Figure 2: X-Point Reconnection Events in Tabletop Configuration', fontsize=16, fontweight='bold', y=1.02)
plt.show()
```
