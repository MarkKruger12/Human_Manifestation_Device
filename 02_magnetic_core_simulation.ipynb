```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.constants import mu_0

def calculate_A_field(radius, current, z_range):
    """
    Calculates the magnitude of the Vector Potential (A) 
    along the central axis (z) of a thin electromagnetic coil.
    """
    # A_phi = (mu_0 * I * R^2) / (4 * (R^2 + z^2)^(3/2)) 
    # Note: This is a simplified axial approximation for comparison
    A = (mu_0 * current * radius**2) / (4 * (radius**2 + z_range**2)**(1.5))
    return A

# Parameters for your apparatus
z = np.linspace(-0.1, 0.1, 500)  # 20cm range
coil_radius = 0.025             # 25mm coil
I_static = 5.0                  # 5 Amps DC
I_pulsed = 5.0 + 0.5 * np.sin(2 * np.pi * 7.83) # Adding Schumann resonance

A_static = calculate_A_field(coil_radius, I_static, z)
A_pulsed = calculate_A_field(coil_radius, I_pulsed, z)

plt.figure(figsize=(10, 6))
plt.plot(z, A_static, label='Static Electromagnet (5A)')
plt.fill_between(z, A_static * 0.95, A_static * 1.05, alpha=0.2, label='A-Field Variance (Schumann Pulse)')
plt.title("Vector Potential (A) Profile: Electromagnetic Pivot")
plt.xlabel("Distance from Center (m)")
plt.ylabel("Potential Magnitude (Webers/m)")
plt.legend()
plt.grid(True)
plt.show()
```
